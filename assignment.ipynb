A549 human lung carcinoma cell line were exposed to X-ray, C ion of LET 13 keV/μm, and C ion of LET 70 keV/μm in aerobic and hypoxic condition. Survival fractions were measured for multiple dose of each radiation types. Standard deviations, SDs are also indicated and are from three independent experiments.

![data](./data.png)

#### A) Draw a set of survival curves exposed in aerobic and hypocix condiction for each radiation types

```python
import matplotlib.pyplot as plt
import math
import numpy as np

Gy_1_aerobic = np.array([0,0.5,1,2,3,5,7,9])
SF_xray_aerobic = np.array([1,0.85,0.744,0.452,0.2224,0.054,0.014,0.00122])
SF_LEF13_aerobic = np.array([1,0.75,0.61,0.34,0.135,0.044,0.0075,0.00064])
Gy_2_aerobic = np.array([0,0.2,0.5,1,2,3,5])
SF_LEF70_aerobic = np.array([1,0.75,0.45,0.286,0.105,0.0306,0.00194])

#plot and fit the survival curves exposed in aerobic condition
plt.plot(Gy_1_aerobic, SF_xray_aerobic, 'bo', label= "X-ray LET 2KeV/um")
plt.plot(Gy_1_aerobic, SF_LEF13_aerobic, 'gs', label = 'C ion LET 13KeV/um')
plt.plot(Gy_2_aerobic, SF_LEF70_aerobic, 'rv', label='C ion LET 70KeV/um')
plt.yscale("log")
p_xray_aerobic = np.polyfit(Gy_1_aerobic, np.log(SF_xray_aerobic), 2)
p_let13_aerobic = np.polyfit(Gy_1_aerobic, np.log(SF_LEF13_aerobic), 2)
p_let70_aerobic = np.polyfit(Gy_2_aerobic, np.log(SF_LEF70_aerobic), 2)


plt.plot(Gy_1_aerobic, np.exp(np.polyval(p_xray_aerobic, Gy_1_aerobic)), color="blue")
plt.plot(Gy_1_aerobic, np.exp(np.polyval(p_let13_aerobic, Gy_1_aerobic)), color="green")
plt.plot(Gy_2_aerobic, np.exp(np.polyval(p_let70_aerobic, Gy_2_aerobic)), color="red")
plt.title("Survival curves in aerobic condition")
plt.legend()
plt.show()
```

```python
Gy_1_hypoxic = np.array([0,1,3,5,7,9,12,15,18])
Gy_2_hypoxic = np.array([0,1,3,5,7,9,12,15])
Gy_3_hypoxic = np.array([0,0.2,0.5,1,2,3,5])
SF_xray_hypoxic = np.array([1,0.9,0.754,0.51,0.35,0.242,0.115,0.051,0.018])
SF_LEF13_hypoxic = np.array([1,0.813,0.65,0.35,0.198,0.098,0.028,0.0055])
SF_LEF70_hypoxic = np.array([1,0.842,0.595,0.405,0.166,0.0496,0.00534])

#plot and fit the survival curves exposed in aerobic condition
plt.plot(Gy_1_hypoxic, SF_xray_hypoxic, 'bo', label= "X-ray LET 2KeV/um")
plt.plot(Gy_2_hypoxic, SF_LEF13_hypoxic, 'gs', label = 'C ion LET 13KeV/um')
plt.plot(Gy_3_hypoxic, SF_LEF70_hypoxic, 'rv', label='C ion LET 70KeV/um')
plt.yscale("log")
p_xray_hypoxic = np.polyfit(Gy_1_hypoxic, np.log(SF_xray_hypoxic), 2)
p_let13_hypoxic = np.polyfit(Gy_2_hypoxic, np.log(SF_LEF13_hypoxic), 2)
p_let70_hypoxic = np.polyfit(Gy_3_hypoxic, np.log(SF_LEF70_hypoxic), 2)


plt.plot(Gy_1_hypoxic, np.exp(np.polyval(p_xray_hypoxic, Gy_1_hypoxic)), color="blue")
plt.plot(Gy_2_hypoxic, np.exp(np.polyval(p_let13_hypoxic, Gy_2_hypoxic)), color="green")
plt.plot(Gy_3_hypoxic, np.exp(np.polyval(p_let70_hypoxic, Gy_3_hypoxic)), color="red")
plt.title("Survival curves in hypoxic condition")
plt.legend()
plt.show()
```
