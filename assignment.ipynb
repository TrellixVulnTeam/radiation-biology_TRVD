A549 human lung carcinoma cell line were exposed to X-ray, C ion of LET 13 keV/μm, and C ion of LET 70 keV/μm in aerobic and hypoxic condition. Survival fractions were measured for multiple dose of each radiation types. Standard deviations, SDs are also indicated and are from three independent experiments.

#### A) Draw a set of survival curves exposed in aerobic and hypocix condiction for each radiation types

```python
import matplotlib.pyplot as plt
import math
import numpy as np
Gy_1 = np.array([0,0.5,1,2,3,5,7,9])
SF_xray = np.array([1,0.85,0.744,0.452,0.2224,0.054,0.014,0.00122])
SF_LEF13 = np.array([1,0.75,0.61,0.34,0.135,0.044,0.0075,0.00064])
Gy_2 = np.array([0,0.2,0.5,1,2,3,5])
SF_LEF70 = np.array([1,0.75,0.45,0.286,0.105,0.00306,0.00194])

#plot and fit the survival curves exposed in aerobic condition
plt.plot(Gy_1, SF_xray, 'bo', label= "X-ray LET 2KeV/um")
plt.plot(Gy_1, SF_LEF13, 'gs', label = 'C ion LET 13KeV/um')
plt.plot(Gy_2, SF_LEF70, 'rv', label='C ion LET 70KeV/um')
plt.yscale("log")
p_xray = np.polyfit(Gy_1, np.log(SF_xray), 2)
p_let13 = np.polyfit(Gy_1, np.log(SF_LEF13), 2)
p_let70 = np.polyfit(Gy_2, np.log(SF_LEF70), 2)
plt.plot(Gy_1, np.exp(np.polyval(p_xray, Gy_1)), color="blue")
plt.plot(Gy_1, np.exp(np.polyval(p_let13, Gy_1)), color="green")
plt.plot(Gy_2, np.exp(np.polyval(p_let70, Gy_2)), color="red")
plt.title("Survival curves in aerobic condition")
plt.legend()
plt.show()
```
